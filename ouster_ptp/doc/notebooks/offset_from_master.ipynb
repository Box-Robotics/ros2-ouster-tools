```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')

import sys
import os

import pandas as pd

sys.path.append('../../')
from ouster_ptp.json.current_data_set import df_from_file
```

```python
OU_ID = "bc0fa7.fffe.000792-1"

df = df_from_file("./ptp_json_log-02.txt")
df = df[df.id == OU_ID]
# convert nanos to micros
df["offset_from_master"] = df["offset_from_master"].astype(float) * 1e-3
df["mean_path_delay"] = df["mean_path_delay"].astype(float) * 1e-3
```

```python
df["recv_stamp_diff"] = (df.recv_stamp - df.recv_stamp.iloc[0])*1e-9

ax = df.plot(x="recv_stamp_diff", 
             y="offset_from_master",
             legend=False, style=['-o'])
ax.set_xlabel("Time (seconds)")
_ = ax.set_ylabel("Offset from Master (usec)")
```

```python
axarr = df.hist(column="offset_from_master", bins=100)
for ax in axarr.flatten():
    ax.set_xlabel("Time (usec)")
```
