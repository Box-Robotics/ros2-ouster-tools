```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')

import sys
import os

import pandas as pd

sys.path.append('../../')
from ouster_ptp.json.sysclk import df_from_file
```

```python
df = df_from_file("./sysclk-01.txt")
df["timediff"] = df.system_time - df.lidar_time
df["timepoint"] = df.system_time - df.system_time.iloc[0]
```

```python
ax = df.plot(x="timepoint", 
             y="timediff",
             legend=False, style=['-'])
ax.set_xlabel("Time (seconds)")
_ = ax.set_ylabel("System Time - Lidar Time (seconds)")
```

```python
axarr = df.hist(column="timediff", bins=50)
for ax in axarr.flatten():
    ax.set_xlabel("Time Diff (sec)")
```
